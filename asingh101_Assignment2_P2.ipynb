```python
import numpy as np
import matplotlib.pyplot as plt

d = 1000
ks = np.arange(10, 1001, 20)

v = np.random.randn(d)
v_norm = np.linalg.norm(v)

ratios = []

for k in ks:
    M = np.random.randn(k, d) / np.sqrt(k)
    Mv = M @ v
    ratio = np.linalg.norm(Mv) / v_norm
    ratios.append(ratio)

plt.figure(figsize=(8,6))
plt.plot(ks, ratios, marker='o', linestyle='-')
plt.xlabel("k (projected dimension)", fontsize=12)
plt.ylabel(r"$\|Mv\| / \|v\|$", fontsize=12)
plt.title("Empirical Study of Random Projection", fontsize=14)
plt.grid(True)
plt.show()
```
